# # Scrape the McDonald's Website for the Prices

```python
browser = Browser('chrome')
```

```python
url = 'https://www.mcdonaldsmenu.ca/mcdonalds-breakfast-menu/'
browser.visit(url)
```

```python
html = browser.html
html_soup = bs(html, 'html.parser')
```

```python
mcds_table = html_soup.find('table', class_='wpdtSimpleTable')
```

```python
mcds_items = []
mcds_prices = []

# Loop through the rows
if mcds_table:
    rows = mcds_table.find_all('tr')
    for row in rows[1:]:  # Skip the header row
        cells = row.find_all('td')
        if len(cells) == 2:
            item = cells[0].text.strip()
            price = cells[1].text.strip()
            mcds_items.append(item)
            mcds_prices.append(price)
else:
    print("Table not found or mcds_table is None.")
```

```python
mcds_data = {
    'Item': mcds_items,
    'Prices': mcds_prices
}
mcds_price_df = pd.DataFrame(mcds_data)
mcds_price_df
```

```python
mask = mcds_price_df['Item'].str.contains('Chicken') | mcds_price_df['Item'].str.contains('BLT')

# Use the boolean mask to exclude rows containing the 'exclude_word'
filtered_price_menu = mcds_price_df[~mask]
filtered_price_menu
```

```python
mcds_price_df.to_csv('breakfast_mcds_prices.csv', index=False, sep='\t', encoding='utf-16')
```

```python
filtered_price_menu.to_csv('breakfast_mcds_prices_filtered.csv', index=False, sep='\t', encoding='utf-16')
```

## Add Calorie, Protein, Total Fat Columns with Sums to Prices Menu

```python
#set the index as the menu item
copy_menu_df = select_columns_df.copy()
copy_menu_df = copy_menu_df.set_index('Item')
copy_menu_df
```

```python
re_indexed_filtered_price_menu = filtered_price_menu.reset_index(drop=True)
re_indexed_filtered_price_menu
```

```python
# Create a list of the different meal categories
meals = [
    ['Hash Brown', 'Sausage Burrito'],
    ['Hash Brown', 'Sausage Burrito', 'Sausage Burrito'],
    ['Hash Brown', 'Bacon, Egg & Cheese Bagel'],
    [],
    ['Hash Brown', 'Bacon, Egg & Cheese Bagel'],
    ['Hash Brown', 'Bacon, Egg & Cheese Bagel'],
    ['Hash Brown', 'Bacon, Egg & Cheese Bagel'],
    ['Hash Brown', 'Egg McMuffin'],
    [],
    ['Hash Brown', 'Sausage, Egg & Cheese McGriddles'],
    ['Hash Brown', 'Sausage McMuffin with Egg'],
    ['Hash Brown', 'Bacon, Egg & Cheese McGriddles'],
    ['Hash Brown', 'Sausage McGriddles'],
    []
    
]

meals_data = []

# Create a new list for each result that is to be calculated
# mcds_price_df['Total Calories - Meal'] = 0
# mcds_price_df['Total Fat (% Daily Value) - Meal'] = 0
# mcds_price_df['Total Sugars - Meal'] = 0
# mcds_price_df['Total Protein - Meal'] = 0

# Loop through each meal to calculate the nutritous value for each
for i, meal in enumerate(meals, start=1):
    total_calories = copy_menu_df.loc[meal, 'Calories'].sum()
    total_fat = copy_menu_df.loc[meal, 'Total Fat (% Daily Value)'].sum()
    total_sugars = copy_menu_df.loc[meal, 'Sugars'].sum()
    total_protein = copy_menu_df.loc[meal, 'Protein'].sum()

# Append the new values to the meal data list
    meals_data.append([total_calories, total_fat, total_sugars, total_protein])

# Create a new data frame with the meals nutritious values
meals_nutrients_df = pd.DataFrame(meals_data, columns=['Total Calories', 'Total Fat (% Daily Value)', 'Total Sugars', 'Total Protein'])

meals_nutrients_df
```

```python
merged_menu_price_df = re_indexed_filtered_price_menu.join(meals_nutrients_df)
merged_menu_price_df
```
